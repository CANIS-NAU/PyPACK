```python
import sys
import pandas as pd
import libpypack
from libpypack.locations import map_locations
from libpypack.locations import webpage_locations
from libpypack.visualization import generate_maps
from libpypack.visualization import choropleth
from libpypack.visualization import heatmap

# Starts Docker to allow for use of location_extraction.
from libpypack.locations.start_docker import run_docker

# If you have already ran this part succesfully, you may comment out the line below.
run_docker()
```

### Predefine Paths for
+ Shapefile - The GeoDataset you wish to parse (i.e. a United States Shapefile)
+ CSV_FILE - The file you wish to extract information from

```python
csv_file = "/Users/thatcher/Documents/GraduateSchool/Thesis/FORKS/PyPACK/notebooks/aggregate_political_file_500.csv"
shapefile = "/Users/thatcher/Documents/GraduateSchool/Thesis/FORKS/PyPACK/libpypack/examples/states_21basic/states.shp"
```

### An Example of what a DataFrame could look like

```python
location_dataframe = pd.read_csv(csv_file, sep="\t")
location_dataframe
```

### Generating/Extracting Files with Locations
 + map_locations.location_df(csv_file):  Returns a DataFrame with extracted locations from text in the 'locs' column in the respected CSV/TSV
 + generate_maps.get_loc_gdf(loc_df): Returns a GeoPandas Dataframe with parsed locations coordinates embedded in the Lattitude/Longitude Columns needed for GeoPandas

```python
# Generate the Dataframe with Locations
loc_df = map_locations.locations_df(csv_file, sep="\t", output_dir="/Users/thatcher/Desktop", df_column="Full_Text")

# Generate the geoDataFrame with Lat/Lon Information
loc_gdf = generate_maps.get_loc_gdf(loc_df)
```

### Heatmap

+ heatmap.heatmap(loc_gdf): Writes to a file labeled 'heatmap.html' to the location defined by 'output_dir'. It also returns a Heatmap defined in Folium (which you can call to see the map)

```python
# Generate the heatmap
hmap = heatmap.heatmap(loc_gdf, output_dir="/Users/thatcher/Desktop")

# Display the heatmap
hmap
```

### Choropleth Map Generation

+ choropleth.choropleth_map(loc_gdf): Returns a GeoPandas Object with additional columns that is needed for plotting this map.

+ choropleth.plot_map(gdf): Writes an image called "choropleth.png" to your respected output_dir and returns a map that is displayable.

```python
# Generate the Choropleth Map
gdf = choropleth.choropleth_map(loc_gdf, shp_path="/Users/thatcher/Documents/GraduateSchool/Thesis/FORKS/PyPACK/libpypack/examples/states_21basic/states.shp")

# Plot the choropleth map
plot = choropleth.plot_map(gdf, output_dir="/Users/thatcher/Desktop")
```

### Generation of Overlay Map
+ generate_maps.generate_overlay_gdf(loc_df): Given a shapefile for the field 'shp_file' and a dataframe with locations added on. It generates a GeoPandas for both the shapefile and location dataframe that it was passed, which you then need to pass on to the next funciton.
+ generate_maps.plot_gdf(gdf, loc_gdf): Returns an image named "overlay.png" to the output_dir you specify and returns a plot that is displayable.

```python
# Generate the Location Overlay Map
gdf, loc_gdf = generate_maps.generate_overlay_gdf(loc_df, shp_path="/Users/thatcher/Documents/GraduateSchool/Thesis/FORKS/PyPACK/libpypack/examples/states_21basic/states.shp")

# Plot the Overlay Map
plot = generate_maps.plot_gdf(gdf, loc_gdf, output_dir="/Users/thatcher/Desktop")
```

### Webscraping URLs
+ webpage_locations.extract_webpage_locations(csv_file): You need to specify column_name to be the column you wish to scrape for content. This returns a datframe and a file named scraped_website_data.csv to the output_dir you specify.
+ webpage_locations.map_web_locations(web_df): Returns another Dataframe with the locations parsed and put into a column named "Web_Locs".

```python
web_df = webpage_locations.extract_webpage_locations(csv_file, sep="\t", output_dir="/Users/thatcher/Desktop", column_name="URLs")
mapped_df = webpage_locations.map_web_locations(web_df, sep="\t", output_dir="/Users/thatcher/Desktop")
```

### Plotting the Web Data
+ generate_maps.get_loc_gdf(mapped_df): Reading in the df with locations mapped (in this case mapped_df). Returns a GeoPandas DataFrame that can be used to generate the same maps as above. Just make sure to change the column to "Web_Locs" just like below!

```python
# Generate the geoDataFrame with Lat/Lon Information
web_gdf = generate_maps.get_loc_gdf(mapped_df, column_name="Web_Locs")
```

```python
# Generate the Location Overlay Map
web_gdf, web_loc_gdf = generate_maps.generate_overlay_gdf(mapped_df, shp_path="/Users/thatcher/Documents/GraduateSchool/Thesis/FORKS/PyPACK/libpypack/examples/states_21basic/states.shp", column_name="Web_Locs")

# Plot the Overlay Mapw
plot = generate_maps.plot_gdf(web_gdf, web_loc_gdf, output_dir="/Users/thatcher/Desktop")

```

```python
# Generate the heatmap
hmap = heatmap.heatmap(web_loc_gdf, output_dir="/Users/thatcher/Desktop")

# Display the heatmap
hmap
```

```python
# Generate the Choropleth Map
new_web_gdf = choropleth.choropleth_map(web_loc_gdf, shp_path="/Users/thatcher/Documents/GraduateSchool/Thesis/FORKS/PyPACK/libpypack/examples/states_21basic/states.shp")

# Plot the choropleth map
plot = choropleth.plot_map(new_web_gdf, output_dir="/Users/thatcher/Desktop")
```
